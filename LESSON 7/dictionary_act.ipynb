### Problem 4 – List Methods: append, insert, remove, sort

### Problem 9 – Iterating with keys(), values(), and items()

Using `contacts_dict` from Problem 8, write three separate loops that demonstrate the three ways to iterate over a dictionary.

**Tasks:**
- **Loop 1:** Print all names (keys) using `for name in contacts_dict`
- **Loop 2:** Print all contact sub-dictionaries (values) using `.values()`
- **Loop 3:** Use `.items()` to print a formatted message for each person:
  ```
  Maria — Phone: 09171234567 | Email: maria@email.com
  ```

````python
contacts_dict = {
    "Maria": {"phone": "09171234567", "email": "maria@email.com"},
    "Juan":  {"phone": "09289876543", "email": "juan@email.com"},
    "Ana":   {"phone": "09351112233", "email": "ana@email.com"},
}

# Loop 1: keys only
print("--- Names ---")
for name in contacts_dict:
    print(name)


# Loop 2: values only
print("--- Contact details ---")
for info in contacts_dict.values():
    print(info)


# Loop 3: .items() — formatted output
print("--- Full directory ---")

for name, info in contacts_dict.items():
    print(f"{name} - Phone: {info['phone']} | Email: {info['email']}")
````

### Problem 10 – Nested Structure: List of Dictionaries

You are given a class roster as a list of dictionaries. Complete each task using the concepts from this chapter.

````python
roster = [
    {"name": "Maria Santos",   "gwa": 1.50, "year": 2},
    {"name": "Juan dela Cruz", "gwa": 2.25, "year": 2},
    {"name": "Ana Reyes",      "gwa": 1.75, "year": 3},
    {"name": "Carlo Bautista", "gwa": 3.00, "year": 2},
    {"name": "Rosa Garcia",    "gwa": 1.25, "year": 3},
]

# Task 1: Print a formatted name | GWA table using a loop
print("Name                | GWA")
print("-" * 28)

for data in roster:
    print(f"{data['name']:<19} | {data['gwa']:.2f}")


# Task 2: Compute class average GWA
# Hint: use a list comprehension to extract GWAs first
print("-" * 28)

gwa_list = [data["gwa"] for data in roster] #list comprehnsion - indepth discussion next meeting
class_average =sum(gwa_list)/len(gwa_list)
print(f"Class average GWA: {class_average}")


# Task 3: Build honor_roll — students with GWA <= 1.75
# Use a list comprehension, then print their names

honors = [data for data in roster if data["gwa"] <= 1.75]
honors_name = ",".join([data["name"] for data in honors])
print(f"Honor roll: {honors_name}")


# Task 4: Find the top student (lowest GWA)
top_student = min(roster, key=lambda data: data["gwa"])
print(f"Top student: {top_student['name']} (GWA: {top_student['gwa']})")

# Task 5: Add "remarks" key to each student dict using a loop
# "Passed" if GWA <= 3.00, otherwise "For Removal"
````
